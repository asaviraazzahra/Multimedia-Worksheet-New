## Apa itu Data Multimedia?

Data multimedia adalah informasi yang dikodekan dalam berbagai format untuk merepresentasikan dunia nyata:

- **Audio (1D)**: Sinyal satu dimensi yang berubah terhadap waktu
  - Contoh: musik, suara, speech
  - Representasi: amplitudo vs waktu
  
- **Gambar (2D)**: Matriks nilai intensitas dalam ruang dua dimensi
  - Contoh: foto, ilustrasi, grafik
  - Representasi: intensitas pixel pada koordinat (x,y)
  
- **Video (2D + Waktu)**: Rangkaian frame (gambar) yang ditampilkan berurutan
  - Contoh: film, rekaman, animasi
  - Representasi: frame berubah terhadap waktu dengan frame rate tertentu

### A4. TODO: Spectrogram log-dB

**Instruksi:**
Hitung STFT dan tampilkan spectrogram dalam skala log-dB:
- Gunakan parameter standar (n_fft=1024, hop_length=256)
- Tampilkan dengan colorbar
- Label sumbu: waktu (detik) dan frekuensi (Hz)

**Analisis yang diperlukan:**
Jelaskan perbedaan informasi yang didapat dari spectrogram dibanding waveform.

```python
# Menghitung STFT dan Menampilkan spectrogram dalam skala log-dB

# Path ke file audio
audio_path = 'data/Golden - Huntrix.mp3'

# Muat audio menggunakan librosa
try:
    audio_data, sample_rate = librosa.load(audio_path, sr=None)  # sr=None untuk mempertahankan sample rate asli

    # Hitung STFT
    n_fft = 1024
    hop_length = 256
    stft = librosa.stft(audio_data, n_fft=n_fft, hop_length=hop_length)
    stft_db = librosa.amplitude_to_db(np.abs(stft), ref=np.max)  # Konversi ke skala log-dB

    # Plot spectrogram
    plt.figure(figsize=(12, 6))
    librosa.display.specshow(stft_db, sr=sample_rate, hop_length=hop_length, x_axis='time', y_axis='hz', cmap='viridis')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram (log-dB)')
    plt.xlabel('Waktu (detik)')
    plt.ylabel('Frekuensi (Hz)')
    plt.show()

except Exception as e:
    print(f"Error saat memuat atau memproses file audio: {e}")
```

**Perbedaan informasi yang didapat dari spectrogram dibanding waveform:**

**Waveform**
- Waveform itu nunjukin seberapa keras atau pelan suara seiring waktu.
- Kita bisa lihat kalo pas intro suaranya masih kecil, pas chorus suaranya rame banget dan padat, pas break/tenang volumenya turun lagi.
- Jadi, waveform tuh fokus ke volume naik-turun sepanjang lagu.
- Kalo diibaratkan kayak grafik suara naik-turun, mirip kalo liat volume bar yang gerak-gerak.

**Spectogram (peta frekuensi audio)**
- Spectrogram itu lebih detail, dia nunjukin nada rendah sampai nada tinggi seiring waktu.
- Warna di spectrogram nunjukin seberapa kuat bunyinya: kuning terang = kenceng, biru/ungu = lemah.
- Dari sini bisa kelihatan, bass itu nada rendah (di bawah 250 Hz), vokal atau instrumen tengah (500‚Äì4000 Hz), efek tinggi (lebih dari 5000 Hz).
- Jadi spectrogram bisa kasih tau nada apa yang main, bukan cuma keras-pelannya.
- Ibaratnya spectrogram tuh kayak peta warna musik, bisa keliatan kapan bass nendang, vokal muncul, atau suara tinggi rame.

### A5. TODO: MFCC

**Instruksi:**
Hitung dan tampilkan minimal 13 koefisien MFCC sebagai heatmap:
- Sumbu X: waktu (frame)
- Sumbu Y: koefisien MFCC (1-13)
- Gunakan colorbar dan judul yang jelas

**Analisis yang diperlukan:**
Interpretasi sederhana: apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?

```python
# Mengimport library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# Path ke file audio
audio_path = 'data/Golden - Huntrix.mp3'

try:
    # Muat audio dan hitung MFCC
    audio_data, sample_rate = librosa.load(audio_path, sr=None)
    mfccs = librosa.feature.mfcc(y=audio_data, sr=sample_rate, n_mfcc=13)

    # Plot menggunakan librosa.display.specshow
    plt.figure(figsize=(12, 6))
    img = librosa.display.specshow(
        mfccs, 
        sr=sample_rate,
        x_axis='time',
        y_axis='mel',
        cmap='viridis'
    )
    plt.colorbar(format='%+2.0f')
    plt.title('MFCC (Mel-Frequency Cepstral Coefficients)')
    plt.xlabel('Waktu (detik)')
    plt.ylabel('MFCC Koefisien')
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Error saat memproses MFCC: {e}")
```

**Apakah pola MFCC stabil atau berubah-ubah? Apa potensi maknanya?**

Jadi, kalau kita lihat grafik MFCC di lagu Golden ‚Äì Huntrix, warnanya tuh keliatan cenderung stabil, alias mirip-mirip sepanjang lagu. Artinya, musiknya punya pola yang konsisten, terutama di bagian beat dan instrumen utama. 

Ada bagian tertentu (misalnya di menit sekitar 1:40 ‚Äì 2:00 dan di akhir) yang warnanya agak berubah dan itu nunjukin adanya transisi musik (bagian break atau outro).

- Intro : masih ringan, belum full beat.

Pola MFCC di lagu ini lebih banyak stabil, karena musik elektronik biasanya repetitif dengan beat yang konsisten. Tapi ada juga momen pola berubah (misalnya di break dan outro), yang nunjukin adanya variasi biar lagu nggak monoton.

### B4. TODO: Histogram Warna

**Instruksi:**
Tampilkan histogram distribusi intensitas untuk channel R, G, B:
- Range: 0-255
- Plot terpisah atau overlay dengan warna sesuai channel
- Label sumbu: intensitas pixel dan frekuensi
- Legend yang jelas

**Analisis yang diperlukan:**
Analisis: channel mana yang dominan? Bagaimana kontras gambar? Seperti apa sebaran intensitasnya?

```python
try:
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Membaca gambar
    img_path = 'data/sunsetbgtyh.jpg'
    img = cv2.imread(img_path)
    
    if img is None:
        raise Exception("Gagal membaca gambar")

    # Memisahkan channel BGR
    b, g, r = cv2.split(img)
    
    # Membuat figure dengan ukuran yang sesuai
    plt.figure(figsize=(12, 6))
    
    # Plot histogram untuk setiap channel
    plt.hist(r.ravel(), bins=256, range=[0, 256], color='red', alpha=0.7, label='Red Channel')
    plt.hist(g.ravel(), bins=256, range=[0, 256], color='green', alpha=0.7, label='Green Channel')
    plt.hist(b.ravel(), bins=256, range=[0, 256], color='blue', alpha=0.7, label='Blue Channel')

    # Menambahkan label dan judul
    plt.xlabel('Intensitas Pixel')
    plt.ylabel('Frekuensi')
    plt.title('Histogram Distribusi Warna RGB')
    plt.legend()
    plt.grid(True, alpha=0.2)
    
    # Mengatur layout
    plt.tight_layout()
    
    # Menampilkan plot
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```

**Analisis Histogram Warna**

1. **Warna yang paling dominan**
   - Dari grafik, warna biru kelihatan paling banyak dan menyebar luas dari gelap sampai terang
   - Warna hijau muncul cukup kuat di bagian terang (sekitar 150‚Äì180)
   - Warna merah lebih banyak ada di bagian gelap dan menengah (sekitar 30‚Äì130).

### C2. TODO: Baca & Metadata

**Instruksi:**
Baca video dengan OpenCV dan tampilkan metadata:
- Resolusi (Width √ó Height)
- Frame rate (fps)
- Jumlah total frame
- Durasi (detik)
- Klasifikasi resolusi (HD, Full HD, 4K, dll.)

**Analisis yang diperlukan:**
Jelaskan pentingnya parameter-parameter tersebut untuk analisis video atau aplikasi tertentu.

```python
try:
    import cv2
    import math

    # Membaca video
    video_path = 'data/vidsunset.mp4'
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        raise Exception("Error: Tidak dapat membuka video")

    # Mengambil metadata video
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps

    # Menentukan klasifikasi resolusi
    def classify_resolution(w, h):
        if w >= 3840 and h >= 2160:
            return "4K Ultra HD"
        elif w >= 2560 and h >= 1440:
            return "2K Quad HD"
        elif w >= 1920 and h >= 1080:
            return "Full HD (1080p)"
        elif w >= 1280 and h >= 720:
            return "HD (720p)"
        else:
            return "Standard Definition"

    # Menampilkan informasi
    print("=== Metadata Video ===")
    print(f"Resolusi: {width} √ó {height} pixel")
    print(f"Frame rate: {fps} fps")
    print(f"Jumlah frame: {total_frames}")
    print(f"Durasi: {duration:.2f} detik ({math.floor(duration/60)}:{int(duration%60)} menit)")
    print(f"Klasifikasi resolusi: {classify_resolution(width, height)}")

    # Menutup video
    cap.release()

except Exception as e:
    print(f"Error: {e}")
```

**Analisis Parameter Video**

1. **Resolusi (Width √ó Height)**

   Ini nunjukin seberapa detail kualitas gambar di video. Makin tinggi resolusi, makin tajam gambarnya. Resolusi ini penting buat nyesuain sama layar yang dipakai, ukuran file, dan seberapa cepat video bisa diproses.

2. **Frame Rate (fps)**

   Ini jumlah gambar per detik yang ditampilin. Makin tinggi fps, makin halus gerakannya. Contohnya:
   - Game biasanya butuh 30‚Äì60 fps biar smooth.
   - Film cukup 24 fps karena itu udah standar sinematik.
   - CCTV kadang pakai 15 fps biar hemat tempat penyimpanan.

3. **Total Frame & Durasi**

   Ini dipakai buat ngitung berapa lama video berjalan, seberapa besar butuh storage, dan berapa lama proses render atau editingnya.

4. **Klasifikasi Resolusi**

   Setiap level resolusi punya fungsi masing-masing:
   - HD (720p): pas buat nonton di HP atau streaming ringan.
   - Full HD (1080p): standar umum buat nonton di laptop/TV.
   - 4K: dipakai buat konten premium atau profesional karena super detail.

### C3. TODO: Tampilkan 3 Frame (Awal‚ÄìTengah‚ÄìAkhir)

**Instruksi:**
Ambil dan tampilkan 3 frame representatif:
- Frame pertama (index 0)
- Frame tengah (index ~total_frame/2)
- Frame terakhir (index total_frame-1)
- **Konversi BGR‚ÜíRGB** sebelum ditampilkan
- Subplot dengan judul frame dan timestamp

**Analisis yang diperlukan:**
Deskripsikan perbedaan visual antar frame dan apa yang dapat dipelajari dari sampel frame ini.

```python
try:
    import cv2
    import matplotlib.pyplot as plt

    # Membaca video
    video_path = 'data/vidsunset.mp4'
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        raise Exception("Error: Tidak dapat membuka video")

    # Mengambil total frame dan fps
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    # Indeks frame yang akan diambil
    frames_idx = [0, total_frames//2, total_frames-1]
    frames = []
    timestamps = []

    # Mengambil frame
    for idx in frames_idx:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if ret:
            # Konversi BGR ke RGB
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame_rgb)
            # Menghitung timestamp
            timestamp = idx / fps
            timestamps.append(f"{int(timestamp//60)}:{int(timestamp%60):02d}")

    # Membuat subplot
    plt.figure(figsize=(15, 5))
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(frames[i])
        plt.title(f'Frame {frames_idx[i]}\nTimestamp: {timestamps[i]}')
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

    # Menutup video
    cap.release()

except Exception as e:
    print(f"Error: {e}")
```

**Analisis Frame Video**

**Frame Awal (Frame 0)**
Di awal video, matahari masih lumayan tinggi jadi pencahayaannya terang banget.

**Frame Tengah**
Masuk ke bagian tengah, cahayanya pelan-pelan berubah. Matahari mulai turun, warna langit juga berubah jadi lebih hangat‚Äîkayak ada campuran kuning-orange.

**Frame Akhir**
Di akhir video, udah jelas banget momen sunset. Matahari makin rendah, cahayanya redup, dan langit berubah dominan orange kemerahan.
